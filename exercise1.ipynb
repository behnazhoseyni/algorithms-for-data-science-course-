Exercise 5.4.1:

In Section 5.4.2 we analyzed the spam farm of Fig. 5.16, where every supporting page links back to the target page. Repeat the analysis for aspam farm in which:<br>
**(a) Each supporting page links to itself instead of to the target page.**
 let y be the unknown PageRank of t
   let us compute the PageRank y of target page t. Its PageRank comes from three sources:
    1. Contribution x from outside, as we have assumed.
    2. (1−β)/n, the share of the fraction 1−β of the PageRank that belongs to t. This amount is negligible and will be dropped to simplify the analysis.
Thus, from (1) and (2) above, we can write y = x <br>

**(b) Each supporting page links nowhere.**<br>
is same as (a)<br>
**(c) Each supporting page links both to itself and to the target page.**<br>
let y be the unknown PageRank of t. We shall solve for y. First, the PageRank of each supporting page is <br>
βy/m + (1−β)/n +β/2(βy/m + (1−β)/n)=(β+1)/2(βy/m + (1−β)/n)
<br>The ﬁrst term represents the contribution from t. The PageRank y of t is taxed, so only βy is distributed to t’s successors. That PageRank is divided equally among the m supporting pages. The second term is the supporting page’s share of the fraction 1−β of the PageRank that is divided equally among all pages on the Web. Now, let us compute the PageRank y of target page t. Its PageRank comes from three sources:
    1. Contribution x from outside, as we have assumed.
    2. β/2 times the PageRank of every supporting page; that is, (β+1)/4(βy/m + (1−β)/n)
    3. (1−β)/n, the share of the fraction 1−β of the PageRank that belongs to t. This amount is negligible and will be dropped to simplify the analysis.
Thus, from (1) and (2) above, we can write y = x +(β+1)/4(βy/m + (1−β)/n)
