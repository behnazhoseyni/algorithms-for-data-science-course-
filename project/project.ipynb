## hierarchical clustering

```python
from scipy.cluster.hierarchy import linkage , dendrogram , fcluster
```

```python
df.target=df['target'].astype(int)
df.data=df.drop(columns='target',axis=1)
```

```python
hierachical=linkage(df.data,method='complete')
dendrogram(hierachical)
plt.show()
```

```python
lables=fcluster(hierachical,250, criterion='distance')
lables
```
